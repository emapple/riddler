```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Arc

%matplotlib inline
```

```python
def circle_func(y, cen=(0, 0), rad=1):
    #returns x
    x0, y0 = cen
    return np.sqrt(rad**2 - (y - y0)**2)
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(7, 7)
circle1 = plt.Circle((-0.4085, 0), 1, color='tomato', alpha=0.5)
circle2 = plt.Circle((0.4085, 0), 1, color='orange', alpha=0.5)
ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_patch(Arc((-0.4085, 0), 0.3, 0.5, theta1=-60, theta2=60, edgecolor='black', lw=1, zorder=100))

ax.plot([-0.4085], [0], color='black', marker='o', ms=4)
ax.plot([0.4085], [0], color='black', marker='o', ms=4)
ax.plot([-0.4085, 0], [0, np.sqrt(1 - 0.4085**2)], color='black')
ax.plot([-0.4085, 0], [0, -np.sqrt(1 - 0.4085**2)], color='black')
ax.plot([0, 0], [np.sqrt(1 - 0.4085**2), -np.sqrt(1 - 0.4085**2)], color='black', ls='--', alpha=0.5)
ax.text(-0.35, 0, r'$\theta$', fontsize=14)
ax.fill_betweenx(np.linspace(-np.sqrt(1 - 0.4085**2), np.sqrt(1 - 0.4085**2), 100),
                 0, circle_func(np.linspace(-np.sqrt(1 - 0.4085**2), np.sqrt(1 - 0.4085**2), 100)) - 0.4085,
                facecolor='none', edgecolor='blue', hatch='/')
ax.text(0.2, 0.2, 'A', fontsize=24)

ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)
frame1.axes.get_yaxis().set_visible(False)
```
